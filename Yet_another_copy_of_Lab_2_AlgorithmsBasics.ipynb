```python
# Exercise 6
def func8(A):
    func8_helper(A, 0, len(A) -1)

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - 1 + 1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
        elif n > 2:
            m = n // 3
            func8_helper(A, i, j - m)
            func8_helper(A, i + m, j)
            func8_helper(A, i, j - m)
        return A
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len

    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

##1.2 Arrays
Programs on 1-d arrays like - sum of elements of array, searching an element inarray, finding minimum and maximum element in array, count the number ofeven and odd numbers in array. For all such programs, also find the timeComplexity, compare if there are multiple methods

```python
# write a function named as sum_arr to find sum of elements in given array
def sum_arr(arr):
    total = 0
    for num in arr:
        total += num
    return total

# Example usage
print(sum_arr([1, 2, 3, 4, 5]))  # Output: 15
```

```python
#write a function named as max_min to find minimum (or maximum) element in an array
def max_min(arr):
    if not arr:
        return None, None  # handle empty array

    minimum = arr[0]
    maximum = arr[0]

    for num in arr:
        if num < minimum:
            minimum = num
        if num > maximum:
            maximum = num

    return minimum, maximum

# Example usage
mn, mx = max_min([3, 1, 5, 2, 4])
print("Minimum:", mn)  # Output: 1
print("Maximum:", mx)  # Output: 5
```

```python
# call sum_arr to find sum of elements in given array for arr = [12, 1234, 45, 67, 1]
arr = [12, 1234, 45, 67, 1]
result = sum_arr(arr)
print(result)  # Output: 1359




```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
def Counting_EvenOdd(arr):
    even_count = 0
    odd_count = 0

    for num in arr:
        if num % 2 == 0:
            even_count += 1
        else:
            odd_count += 1

    return even_count, odd_count

# Example usage
arr = [1, 2, 3, 4, 5, 6]
even, odd = Counting_EvenOdd(arr)
print("Even count:", even)  # Output: 3
print("Odd count:", odd)    # Output: 3
```

```python
# write a function named as linear_Search to perform a linear search in 1D-array
def linear_Search(arr, target):
    """
    Performs linear search for 'target' in 'arr'.
    Returns the index of target if found, else returns -1.
    """
    for i in range(len(arr)):
        if arr[i] == target:
            return i  # return the index where target is found
    return -1  # target not found

# Example usage
arr = [10, 20, 30, 40, 50]
target = 30
index = linear_Search(arr, target)

if index != -1:
    print(f"Element {target} found at index {index}")
else:
    print(f"Element {target} not found in the array")
```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
def Binary_Search(arr, target):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1

arr = [1, 3, 5, 7, 9, 11]
target = 7

index = Binary_Search(arr, target)

if index != -1:
    print(f"Element {target} found at index {index}")
else:
    print(f"Element {target} not found in the array")
```
